```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
path = '/media/anmol/Study/GITHUB/Workshop_NIT_Meghalaya/Datasets/births.csv'
```

```python
df = pd.read_csv(path)
df = df.dropna()
#sum of total births in a year
df1 = pd.DataFrame(df['year'])
df1 = df1.drop_duplicates()
df1['total_births_per_year'] = df.groupby(['year'])['births'].transform(sum)
df1
```

```python
plt.bar(df1['year'],df1['total_births_per_year'])
plt.xlabel('Births')
plt.ylabel('Year')
plt.title('Bar_Plot')
plt.show()
df1.plot('year','total_births_per_year', kind = 'bar')
```
